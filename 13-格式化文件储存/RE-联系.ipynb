# RE使用大致步骤
- 使用compile将表示正则的字符串编译为一个pattern对象
- 通过pattern对象提供一系列方法对文本进行查找匹配，获得匹配结果，一个Match对象
- 最后使用Match对象提供的属性和方法获得信息
# Re常用函数
- group():获得一个或则多个分组匹配的字符串，当要获得整个匹配的子串时，直接使用group或者group(0)
- start:获取分组匹配的子串在整个字符串中起始位置，参数默认为0
- end: 获取分组匹配的子串在整个字符串的结束位置，默认为0
- span：返回的结构技术(start(group),end(group))


```python
#导入相关包
import re

# 查找数字
#r表示字符串不转义
p = re.compile(r'\d+')
#在字符串"one12twothree33456four78"中进行查找，按照规则p制定的正则进行查找
#返回结果是None表示没有找到，否则会返回Match对象
m = p.match("one12twothree33456four78")

print(m)
```

```python
#导入相关包
import re

# 查找数字
#r表示字符串不转义
p = re.compile(r'\d+')
#在字符串"one12twothree33456four78"中进行查找，按照规则p制定的正则进行查找
#返回结果是None表示没有找到，否则会返回Match对象
#参数3,6表示在字符串中查找的范围
m = p.match("one12twothree33456four78", 3, 6)

print(m)

#上述代码说明的问题
#1.Match可以输入参数表示起始位置
#2.查找到的结果只包含一个，表示第一次进行匹配成功的内容
```

```python
print(m[0])
#获取分组匹配的子串在整个字符串中起始位置
print(m.start(0))
print(m.end(0))
```

```python
import re
# I 表示忽略掉大小写
p = re.compile(r'([a-z]+) ([a-z]+)', re.I)

m = p.match("i am really love python")
print(m)

```

```python
print(m.group(0))
print(m.start(0))
print(m.end(0))
```

```python
print(m.group(1))
print(m.start(1))
print(m.end(1))
```

```python
print(m.groups())
```

# 查找
- search(str, [, pos[, endpos]]):在字符串中查找匹配，pos和endpos表示起始位置
- findall：查找所有
- finditer：查找，返回一个iter结果

```python
import re
p = re.compile(r'\d+')

m = p.search('one12twothree33456four78')

print(m.group())
```

```python
rst = p.findall('one12twothree33456four78')
print(type(rst))
print(rst)
```

# sub替换
- sub(rep1, str[, count])

```python
# sub替换的案例
import re 

p = re.compile(r'(\w+) (\w+)')

s = "hello 123 python 456 love i love"

rst = p.sub(r'hello world', s)

print(rst)
```

# 匹配中文
- 大部分中文表示范围[u4e00-u9fa5],不包括全角中文
- 

```python
import re 

#此处的u表示uUnicode编码
title = u'王志鹏, hello!'

#r前面有时会加u，表示Unicode编码
p = re.compile(r'[\u4e00-\u9fa5]+')
rst = p.findall(title)

print(rst)
```

# 贪婪和非贪婪
- 贪婪：尽可能多的匹配,(*)表示贪婪匹配
- 非贪婪：找到符合条件的最小内容即可，(?)表示非贪婪
- 正则默认使用贪婪匹配

```python
import re

title = '<div>name</div><div>age</div>'

p1 = re.compile(r'<div>.*</div>')
p2 = re.compile(r'<div>.*?</div>')

m1 = p1.search(title)
print(m1.group())
m2 = p2.search(title)
print(m2.group())
```
